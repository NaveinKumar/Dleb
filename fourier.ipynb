```python
# Fourier "Blindfold" Demo  find hidden rhythms in noisy data
import numpy as np
import matplotlib.pyplot as plt

# 1 Create a fake "market" signal
# Time = 0  10 seconds
t = np.linspace(0, 10, 1000)

# Hidden rhythms: one slow (0.5 Hz) + one fast (3 Hz)
signal = 1.2*np.sin(2*np.pi*0.5*t) + 0.6*np.sin(2*np.pi*3*t)

# Add random noise (looks messy to the eye)
noisy_signal = signal + 0.4*np.random.randn(len(t))

# 2 "Blindfolded" Fourier Transform
freq = np.fft.fftfreq(len(t), d=t[1]-t[0])     # frequency axis
fft_vals = np.fft.fft(noisy_signal)
amplitude = np.abs(fft_vals)

# 3 Plot what you "see" vs what Fourier "hears"
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(t, noisy_signal, color='teal')
plt.title("  What we see (noisy market signal)")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(freq[:len(freq)//2], amplitude[:len(freq)//2], color='orange')
plt.title("  What Fourier hears (hidden rhythms)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```
