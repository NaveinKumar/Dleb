```python
# Equation of a Circle in Python
# Circle: x^2 + y^2 = 1 (Unit Circle)

import numpy as np
import matplotlib.pyplot as plt

# Generate theta values between 0 and 2π
theta = np.linspace(0, 2 * np.pi, 100)

# Calculate x and y coordinates for the circle
x = np.cos(theta)
y = np.sin(theta)

print(x)
print(y)

# Plot the circle
plt.plot(x, y, label="x² + y² = 1")

# Set equal aspect ratio so the circle isn’t distorted
plt.gca().set_aspect('equal', adjustable='box')

# Set plot limits
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

# Add title, labels, and grid
plt.title("Unit Circle Centered at Origin")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
