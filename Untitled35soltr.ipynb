```python
import os, json, base58
from solders.keypair import Keypair

phantom_secret = os.environ.get("PHANTOM_SECRET")

if phantom_secret is None:
    raise ValueError("PHANTOM_SECRET not found. Please add it in Colab > Tools > Secrets.")

# If it's a JSON array
try:
    secret_array = json.loads(phantom_secret)
    keypair = Keypair.from_bytes(bytes(secret_array))
except json.JSONDecodeError:
    # If it's a Base58 string
    keypair = Keypair.from_bytes(base58.b58decode(phantom_secret))

print("Wallet Public Key:", keypair.pubkey())
```
