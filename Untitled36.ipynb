```python
# ====================== Install dependencies ======================
!pip install solders solana spl-token-py base58

# ====================== Imports ======================
from solana.rpc.api import Client
from solders.keypair import Keypair
from solders.pubkey import Pubkey
from solders.transaction import Transaction
from solders.instruction import Instruction, AccountMeta
from solders.hash import Hash
import base58
import struct
from google.colab import userdata
import time

# ====================== 1. Connect to Mainnet ======================
client = Client("https://api.mainnet-beta.solana.com")

# ====================== 2. Your Data ======================
MINT_ADDRESS = Pubkey.from_string("mnt1QvurQvZmxhaZaH9e7YnUsRMk1L6CEhsYeC6J959")
TOKEN_2022_PROGRAM_ID = Pubkey.from_string("TokenzQdBNbLqP5VEhdkAS6EPFLC1PHnBqCXEpPxuEb")

SENDER_WALLET = Pubkey.from_string("GTxd5p5AzAHEGX7551X9X5uLLFUmecvBkMYgkQMXQEfN")
RECEIVER_WALLET = Pubkey.from_string("3Xqqry6T9nd3a9NTLeaGfBHzcDsTbMBvR9j6gPHHmods")

# ====================== 3. Load sender keypair ======================
phantom_secret = userdata.get('PHANTOM_SECRET')
keypair = Keypair.from_bytes(base58.b58decode(phantom_secret))
print("Sender wallet pubkey:", keypair.pubkey())

# ====================== 4. Derive Token-2022 ATA addresses ======================
# Token-2022 ATA program ID
ATA_PROGRAM_ID = Pubkey.from_string("ATokenGPvbdGVxr1b2hvZbsiqW5xWH25efTNsLJA8knL")

def get_associated_token_address(owner: Pubkey, mint: Pubkey):
    from solders.pubkey import Pubkey
    seeds = [
        bytes(owner),
        bytes(TOKEN_2022_PROGRAM_ID),
        bytes(mint)
    ]
    return Pubkey.find_program_address(seeds, ATA_PROGRAM_ID)[0]

sender_ata = get_associated_token_address(SENDER_WALLET, MINT_ADDRESS)
receiver_ata = get_associated_token_address(RECEIVER_WALLET, MINT_ADDRESS)
print("Sender ATA:", sender_ata)
print("Receiver ATA:", receiver_ata)

# ====================== 5. Helper: check if ATA exists ======================
def ata_exists(ata_pubkey: Pubkey):
    resp = client.get_account_info(ata_pubkey)
    return resp.value is not None

# ====================== 6. Helper: create ATA if missing ======================
from spl.token.instructions import create_associated_token_account

def create_ata(ata_pubkey: Pubkey, owner: Pubkey):
    if not ata_exists(ata_pubkey):
        print(f"Creating ATA for {owner}...")
        # Build transaction
        tx = Transaction.new_signed_with_payer(
            [create_associated_token_account(
                payer=keypair.pubkey(),
                owner=owner,
                mint=MINT_ADDRESS,
                program_id=TOKEN_2022_PROGRAM_ID
            )],
            payer=keypair.pubkey(),
            signing_keypairs=[keypair],
            recent_blockhash=client.get_latest_blockhash().value.blockhash
        )
        sig = client.send_raw_transaction(bytes(tx)).value
        print(f"ATA created. Tx signature: {sig}")
        # Wait a bit for the network to update
        time.sleep(10)

# Ensure both ATAs exist
create_ata(sender_ata, SENDER_WALLET)
create_ata(receiver_ata, RECEIVER_WALLET)

# ====================== 7. Transfer tokens ======================
amount = 5 * (10 ** 9)  # 5 tokens with 9 decimals
decimals = 9

# Build TransferChecked instruction (manual layout)
instruction_data = struct.pack("<BQB", 12, amount, decimals)  # u8, u64, u8

instruction = Instruction(
    program_id=TOKEN_2022_PROGRAM_ID,
    accounts=[
        AccountMeta(pubkey=sender_ata, is_signer=False, is_writable=True),
        AccountMeta(pubkey=MINT_ADDRESS, is_signer=False, is_writable=False),
        AccountMeta(pubkey=receiver_ata, is_signer=False, is_writable=True),
        AccountMeta(pubkey=SENDER_WALLET, is_signer=True, is_writable=False)
    ],
    data=instruction_data
)

# Get latest blockhash
latest_blockhash = client.get_latest_blockhash().value.blockhash

# Build transaction
tx = Transaction.new_signed_with_payer(
    [instruction],
    payer=SENDER_WALLET,
    signing_keypairs=[keypair],
    recent_blockhash=latest_blockhash
)

# Send transaction
response = client.send_raw_transaction(bytes(tx))
sig = response.value
print("Transaction Signature:", sig)
print(f"View on Solscan: https://solscan.io/tx/{sig}")
```
